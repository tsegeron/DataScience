```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/ab-test.csv').drop(columns='uid')
df.head()
```

```python
df.info()
```

```python
plt.rcParams.update({'font.size': 11})
fig, ax = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
fig.suptitle('Lag before lab deadlines. After and before the Newsfeed', fontsize=15)

ax[0].boxplot([df[(df.time=='after') & (df.group=='control')]['diff'],
              df[(df.time=='after') & (df.group=='test')]['diff']],
              labels=['control', 'test'],
              showmeans=True,
              medianprops={'color':'r', 'linewidth':2},
              boxprops={'color':'g', 'linewidth':3},
              whiskerprops={'color':'r'},
              capprops={'color':'r'}
             )
ax[1].boxplot([df[(df.time=='before') & (df.group=='control')]['diff'],
              df[(df.time=='before') & (df.group=='test')]['diff']],
              labels=['control', 'test'],
              showmeans=True,
              medianprops={'color':'r', 'linewidth':2},
              boxprops={'color':'g', 'linewidth':3},
              whiskerprops={'color':'r'},
              capprops={'color':'r'}
             )
ax[0].grid()
ax[0].set_title('after')
ax[0].set_xlabel('[group]')
ax[1].grid()
ax[1].set_title('before')
ax[1].set_xlabel('[group]')

plt.show()
```
